```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import random
```

# Part 1

```python
x = np.arange(-5, 5, 0.1)
def func(x, a1, a2):
    return a1 * x**2 + a2 * x
y = func(x, 1, 1)
```

```python
plt.plot(x, y, "rx")
```

```python
curveplot = curve_fit(func, x, y)

plt.plot(x, y, "rx")
plt.plot(x, func(x, *curveplot[0]), "b-")
```

```python
data = np.loadtxt("HDalpha_data")
```

```python
x2 = []
y2 = []

for dataval in data:
    x2.append(dataval[0])
    y2.append(dataval[1]) 
    
x2 = np.asarray(x2)
y2 = np.asarray(y2)
```

```python
plt.plot(x2,y2) 
```

```python
def func(x, a1, a2, a3, a4):
    val = a1 * x**4 + a2 * x**3 + a3 * x**2 + a4 * x
    return val/100

curveplot2 = curve_fit(func, x2, y2)

plt.plot(x2, y2, "rx")
plt.plot(x2, func(x2, *curveplot2[0]), "b-")
plt.suptitle('Quartic Function Fit To Given Data', fontsize=16)
```

# Part 2

```python
datapoints1 = []
datapoints2 = []

for x in range(100):
    datapoints1.append(random.random())
    datapoints2.append(random.random())

datapoints1 = np.asarray(datapoints1)
datapoints2 = np.asarray(datapoints2)
```

```python
f, axarr = plt.subplots(2, 2)
axarr[0, 0].scatter(datapoints1[1:26], datapoints2[1:26], color = 'purple', alpha = 0.25, s = 100*datapoints2[1:26])
axarr[0, 1].scatter(datapoints1[26:51], datapoints2[26:51], color = 'orange', alpha = 0.5, s = 100*datapoints1[26:51])
axarr[1, 0].scatter(datapoints1[51:76], datapoints2[51:76], color = 'turquoise', alpha = 0.75, s = 100*datapoints2[51:76])
axarr[1, 1].scatter(datapoints1[76:101], datapoints2[76:101], color = 'magenta', alpha = 1, s = 100*datapoints1[76:101])

axarr[0, 0].tick_params(axis='x', bottom='off'), axarr[0, 0].tick_params(axis='y', left='off')
axarr[0, 1].tick_params(axis='x', bottom='off'), axarr[0, 1].tick_params(axis='y', left='off')
axarr[1, 0].tick_params(axis='x', bottom='off'), axarr[1, 0].tick_params(axis='y', left='off')
axarr[1, 1].tick_params(axis='y', left='off'), axarr[1, 1].tick_params(axis='x', bottom='off')

axarr[0,0].set_title("Points 1 to 25")
axarr[0,1].set_title("Points 51 to 75")
axarr[1,0].set_title("Points 26 to 50")
axarr[1,1].set_title("Points 76 to 100")

f.tight_layout()
f.suptitle("Correlation of 2 Sets of Data Points", y = 1.08, fontsize=16)


f.savefig('Part 2 - Plot.pdf', dpi=100)

```

# Part 3

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
xelec = []
yelec = []
zelec = []
XArr = []
YArr = []

for x in range(64):
    xelec.append(random.randint(-32,32))
    yelec.append(random.randint(-32,32))
    zelec.append(random.randint(-32,32))
    
for x in range(64):
    XArr.append(random.randint(-32,32))
    YArr.append(random.randint(-32,32))
    
xelec = np.asarray(xelec)
yelec = np.asarray(yelec)
zelec = np.asarray(zelec)
XArr = np.asarray(XArr)
YArr = np.asarray(YArr)

fluxval = flux_for_dipole(xelec, yelec, zelec, XArr, YArr)

plt.contour(XArr, YArr, fluxval, 30)
plt.title("Contour Plot of Electric Field")
```
